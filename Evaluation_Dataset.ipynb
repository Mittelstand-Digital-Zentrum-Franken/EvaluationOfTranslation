# Analyse der Datensätze

### Installationen und Imports

```python
# Installationen und Imports
import pandas as pd
import os
import json
```

### Laden der JSON-Files

```python
# Laden eine Files
with open(r'data.json', encoding="utf-8-sig") as file:
    obj = json.load(file)

# Struktur überprüfen und in DataFrame konvertieren
df = pd.json_normalize(obj)
```

### Strukturelle Analyse der JSON Files

```python
# Analysieren der File Struktur und Inhalte
# Datentypen und Keys
# Erste Hierarchie
print("Datentypen und Keys: JSON")
print(f"Datentyp des JSON: {type(obj)}")
print(f"Datentypen der Elemente der JSON: {type(obj[0])}")
print(f"Anzahl der Dictionaries in der JSON: {len(obj)}")
print(f"Keys der Dictionaries: {obj[0].keys()}")

# Datentypen und Keys: Translations
print("")
print("Datentypen und Keys: Translations")
print(f"Datentypen: {type(obj[0]['Translations'])}")
print(f"Anzahl der Dictionaries in der JSON: {len(obj[0]['Translations'])}")
print(f"Keys der Dictionaries: {obj[0]['Translations'][0].keys()}")
```

```python
# Definition einer Funktion zur Ermittlung der Sprachen
def extract_languages(translation):
    languages = []
    number_language = range(len(translation))
    for items in number_language:
        language = translation[items]['Language']
        languages.append(language)
    return languages

# Ermittlung der Sprachen in den JSONs
translation = obj[0]['Translations']
language = extract_languages(translation)

# Ausgabe
print(f"Anzahl der Sprachen: {len(obj[0]['Translations'])}")
print(f"Erste 5 Sprachen: {language[:5]}")
```

```python
# Pfad der JSONS ermitteln
folder = "/Data/atticus"
list_files = os.listdir(folder)

print(f"Anzahl der Files: {len(list_files)}")
```

### Zusammentragen der englischen Texte

```python
# Leere Liste für deutsche Texte erzeugen
list_english_text = []

# Laden der Files und Extraktion der englischen Texte (Source Language = en-US)
for path in list_files:
    path_complete = folder + "//" + path
    with open(path_complete,'r', encoding="utf-8-sig") as file:
        obj = json.load(file)

    # Iterieren durch die Texte der JSON und Selection der englischen Texte
    for text in obj:
        list_english_text.append(text['Text'])

# Ausgabe
list_english_text[:5]
```

### Zusammentragen der deutschen Texte

```python
# Leere Liste für deutsche Texte erzeugen
list_german_text = []

# Laden der Files und Extraktion der deutschen Texte
for path in list_files:
    path_complete = folder + "//" + path
    with open(path_complete,'r', encoding="utf-8-sig") as file:
        obj = json.load(file)

    # Iterieren durch die Texte der JSON
    for text in obj:
        # Überprüfen, ob die Sprache "de-DE" ist und Speichern in einer Liste
        for translation in text['Translations']:
            if translation['Language'] == 'de-DE':
                list_german_text.append(translation['Text'])

# Ausgabe
list_german_text[:5]
```
